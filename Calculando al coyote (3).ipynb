# Calculando al coyote


>Para poder calcular el problema que nos pide el ángulo con la cual el cuerpo pierde contacto con la esquina en el punto (0) y empieza a caer
>necesitamos realizar un diagrama de cuerpo libre para ver de mejor manera las fuerzas que interactuan en el cuerpo

## Datos

>lado b=3[m]

>masa m=2000[kg] 

> coeficiente de friccion µ=0.3

>Suponga que la piedra rueda sin deslizar.

## Diagrama de cuerpo libre

![title](dcl.jpeg)


$$\Sigma F_{x}=0 $$
$$-fr*cos(\theta)+N*sen(\theta)=0$$          como $$fr=u*N$$
$$-u*N*cos(\theta)+N*sen(\theta)=0$$
$$N*(sen(\theta)-u*cos(\theta))=0$$
> Para encontrar el angulo $\theta$ debemos calcular al momento que N=0 ya que justo en ese momento pierde contacto
$$sen(\theta)-u*cos(\theta)=0$$
$$sen(\theta)=u*cos(\theta)$$
$$\frac{sen(\theta)}{cos(\theta)} = u$$
$$tan(\theta)=0.3$$
> despejando el angulo $\theta$ obtenemos
$$\theta=16.70^°$$

>Ahora para poder graficar la variacion de la normal con respecto al angulo de giro tenemos que:
$$\Sigma F_{y}=0$$
$$fr*sin(\theta)+N*cos(\theta)-mg=0$$
$$u*N*sin(\theta)+N*cos(\theta)=mg$$
$$N(u*sin(\theta)+cos(\theta))=2000 [kg] * 9.8[\frac{m}{s^2}]$$
$$N=\frac{19600}{u*sin(\theta)+cos(\theta)}$$

```python
th=np.arange(0,2,1)
vth=np.array(np.deg2rad(th))
N=(19600)/(np.cos(th)+0.3*np.sin(th))

fig=px.line(x=th,y=N,labels={"x":"theta [deg]","y":"Fuerza normal [N]"}, title=" variación de la normal con respecto al ángulo de giro")
fig.show()
```
